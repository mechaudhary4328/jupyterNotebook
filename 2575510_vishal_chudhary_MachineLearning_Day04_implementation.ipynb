```python
x=data.iloc[:,:-1]
y=data.iloc[:,-1:]
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score 
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size=0.2, random_state=1)

print('**********Training Data Shape*************\n')

print('Training data-X- Shape:\t', xtrain.shape)

print()

print('Training data-y- Shape:\t', ytrain.shape)

print()

print(" ***Testing Data Shape***")

print('Testing data (x-Input) Shape:\t', xtest.shape)

print()

print('Testing data(y-Outcome) Shape:\t', ytest.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# Create a KNN classifier with n_neighbors=3
knn_model = KNeighborsClassifier(n_neighbors=3)

print('**********KNN Classifier Model**********')
print('Training Phase\n')

# Train the KNN model on the training data
knn_model.fit(xtrain, ytrain)

print('Training is completed\n')

print('Testing Phase\n')

# Predict the class labels on the testing data
ypred = knn_model.predict(xtest)

print('Predicted Class Labels are:\n', ypred)
print('Testing is also done')
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Assuming you have already predicted ypred

# Calculate and print the accuracy score
print('Accuracy Score:\t', accuracy_score(ytest, ypred))

print()

# Generate and print the confusion matrix
print('Confusion Matrix:\n', confusion_matrix(ytest, ypred))

print()

# Generate and print the classification report
print('Classification Report:\n', classification_report(ytest, ypred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# List of different K values to try
k_values = [3, 5, 7, 9, 11]

# Dictionary to store results for each K value
results = {}

for k in k_values:
    # Create a KNN classifier with the current K value
    knn_model = KNeighborsClassifier(n_neighbors=k)
    
    # Train the KNN model on the standardized training data
    knn_model.fit(xtrain, ytrain)
    
    # Predict the class labels on the standardized testing data
    ypred = knn_model.predict(xtest)
    
    # Calculate the accuracy score for this K value
    accuracy = accuracy_score(ytest, ypred)
    
    # Store the accuracy in the results dictionary
    results[k] = accuracy

# Find the K value with the highest accuracy
best_k = max(results, key=results.get)

print("Accuracy for Different K Values:")
for k, accuracy in results.items():
    print(f"K = {k}: Accuracy = {accuracy}")

print(f"The best K value is K = {best_k} with an accuracy of {results[best_k]}")
```
