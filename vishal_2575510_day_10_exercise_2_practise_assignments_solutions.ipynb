```python
#which three years of the 1930s sqe the most film releases.


# Read the titles DataFrame
file_path = r'D:\Practise exercise\python S1\titles.csv'
titles = pd.read_csv(file_path)

# Filter movies released in the 1930s
movies_1930s = titles[(titles['year'] >= 1930) & (titles['year'] < 1940)]

# Count the number of movies released in each year
movie_counts_by_year = movies_1930s['year'].value_counts()

# Get the top three years with the most film releases
top_3_years = movie_counts_by_year.head(3)

# Display the result
print(top_3_years)
```

```python
#plot the number of films that have been released each decade over the history of cinema

import matplotlib.pyplot as plt

# Read the titles DataFrame
file_path = r'D:\Practise exercise\python S1\titles.csv'
titles = pd.read_csv(file_path)

# Extract the decade from the 'year' column
titles['decade'] = (titles['year'] // 10) * 10

# Group the data by decade and count the number of films
decade_counts = titles['decade'].value_counts().sort_index()

# Plot the data
plt.figure(figsize=(10, 6))
plt.bar(decade_counts.index, decade_counts.values, width=8, align='center', tick_label=decade_counts.index)
plt.xlabel('Decade')
plt.ylabel('Number of Films Released')
plt.title('Number of Films Released Each Decade')
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#which actor or actoress apperar in the most movies in the yera 1945.


# Read the cast DataFrame
file_path = r'D:\Practise exercise\python S1\cast.csv'
cast = pd.read_csv(file_path)

# Filter for movies released in 1945
movies_1945 = cast[cast['year'] == 1945]

# Group by actor's name and count appearances
actor_appearances = movies_1945['name'].value_counts()

# Get the actor or actress with the most appearances in 1945
most_appearances_actor = actor_appearances.idxmax()
num_appearances = actor_appearances.max()

# Display the result
print(f"The actor/actress with the most appearances in 1945 is {most_appearances_actor} with {num_appearances} appearances.")
```

```python
#Plot the n value of the role that judi dench has played over her career.

import matplotlib.pyplot as plt

# Read the cast DataFrame
file_path = r'D:\Practise exercise\python S1\cast.csv'
cast = pd.read_csv(file_path)

# Filter for roles played by Judi Dench
judi_dench_roles = cast[cast['name'] == 'Judi Dench']

# Filter out rows with NaN values in the 'n' column
judi_dench_roles = judi_dench_roles.dropna(subset=['n'])

# Plot the 'n' values over the years
plt.figure(figsize=(12, 6))
plt.plot(judi_dench_roles['year'], judi_dench_roles['n'], marker='o', linestyle='-', color='b')
plt.xlabel('Year')
plt.ylabel('n Value of Roles Played by Judi Dench')
plt.title('n Value of Roles Played by Judi Dench Over Her Career')
plt.grid(True)
plt.tight_layout()

# Show the plot
plt.show()
```

```python
#How many leading (n=1) roles were avilable  to actors,and how many actresses in 1950's. 


# Read the cast DataFrame
file_path = r'D:\Practise exercise\python S1\cast.csv'
cast = pd.read_csv(file_path)

# Filter for roles with n=1 in the 1950s for actors
leading_roles_actors_1950s = cast[(cast['year'] >= 1950) & (cast['year'] <= 1959) & (cast['n'] == 1) & (cast['type'] == 'actor')]

# Filter for roles with n=1 in the 1950s for actresses
leading_roles_actresses_1950s = cast[(cast['year'] >= 1950) & (cast['year'] <= 1959) & (cast['n'] == 1) & (cast['type'] == 'actress')]

# Count the number of leading roles for actors and actresses
num_leading_roles_actors = len(leading_roles_actors_1950s)
num_leading_roles_actresses = len(leading_roles_actresses_1950s)

# Display the results
print(f"Number of leading roles (n=1) available to actors in the 1950s: {num_leading_roles_actors}")
print(f"Number of leading roles (n=1) available to actresses in the 1950s: {num_leading_roles_actresses}")
```
